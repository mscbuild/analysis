Cleaning Data to prepare for analysis
Inspected the newly created data frame to understand the structure and data types. I have double-checked the data structure, data format, and data type

```r
## Inspected the newly created data frame
head(rides24_1)
tail(rides24_1)
glimpse(rides24_1)
str(rides24_1)
summary(rides24_1)
```

Checked for duplicates and removed empty rows and columns

```r
## Checked for duplicates
get_dupes(rides24_1, ride_id)

## Removing empty rows and columns
remove_empty(rides24_1)
```

Checked for the unique value:

```r
unique(rides24_1$rideable_type)
```

```r
unique(rides24_1$member_casual)
```

Added new columns

```r
## Added column for ride_length (ride duration in minutes)
rides24_1$ride_length <- difftime(rides24_1$ended_at,rides24_1$started_at, units = "mins")

## Added column for start_time (Extracting time from date & time column)
rides24_1$start_time <- format(as.POSIXlt(rides24_1$started_at),
                                 format = "%H:%M:%S")
## Added column for start_hour
rides24_1 <- rides24_1 %>%
  mutate(start_hour = hour(hms(rides24_1$start_time)))

## Added column for date
rides24_1$date <- as.Date(rides24_1$started_at)

## Added column for month
rides24_1$month <- format(rides24_1$date, "%B")

## Adding column for day_of_week
rides24_1$day_of_week <- format(as.Date(rides24_1$date), "%A")
```

Removed Bad Data

```r
## Checked for ride_length equal to zero
rides24_1_0 <- rides24_1 %>%
  filter(ride_length == 0)

## Removed rows with ride_length equal to zero
rides24_1 <- rides24_1[!(rides24_1$ride_length == 0),]
```

Converting ride_length into numeric

```r
rides24_1$ride_length <- as.numeric(rides24_1$ride_length)
```

Comparing average, median, maximum, and minimum ride duration by each day of the week for members vs casual users

```r
## Comparing by each day of week for member vs casual
rides24_1 %>%
  group_by(member_casual, day_of_week) %>%
  summarise(number_of_rides = n(),
            average_ride_duration = mean(ride_length),
            median = median(ride_length),
            max = max(ride_length),
            min = min(ride_length)) %>%
  arrange(member_casual, day_of_week)
```

The average ride time by each day for members vs casual users

```r
aggregate(rides24_1$ride_length ~ rides24_1$member_casual + rides24_1$day_of_week, FUN = mean)
```

```r
## Comparing day of week
aggregate(rides24_1$day_of_week ~ rides24_1$member_casual, FUN = max)
```

Comparing member vs casuals by day of week distribution

```r
rides24_1 %>%
  group_by(day_of_week) %>%
  dplyr::summarise(total_count = length(ride_id),
            perc = (length(ride_id) / nrow(rides24_1)) * 100,
            members_perc = (sum(member_casual == "member") / length(member_casual)) * 100,
            casual_perc = (sum(member_casual == "casual") / length(member_casual)) * 100,
            Member_vs_Casual_Perc_Difer = members_perc - casual_perc)
```

```r
## Comparing month
aggregate(rides24_1$month ~ rides24_1$member_casual, FUN = max)
```

**Top 10 stations**

Top 10 Start stations and codes used for both casual and member riders respectively

```r
rides24_1 %>% #Top 10 Start stations for Casual Riders
  group_by(member_casual,start_station_name) %>%
  dplyr::summarise(number_of_ride = n()) %>%
  filter(start_station_name != "", "casual"== member_casual) %>%
  arrange(-number_of_ride) %>%
  head(n=10) %>%
  select(-member_casual)
```

```r
top_station <- rides24_1 %>%
  group_by(start_station_name) %>%
  summarise(total_count = length(na.omit(start_station_name))) %>%
  arrange(desc(total_count)) %>%
  top_n(10)

top_station
```

```r
write.csv(top_station ,"top_station.csv",row.names = T)
```

```r
## Created viz for top 10 stations
ggplot(data  = top_station) +
  geom_col(mapping = aes(x = start_station_name, y = total_count,
                         fill = start_station_name), position = "dodge") +
  labs(title = "Top Ten Start Station") +
  theme(axis.text.x = element_text(angle = 90))
```

```r
## Top Ten Stations for casual Riders
top_station2 <- rides24_1 %>%
  filter(member_casual == "casual")

top_station_casual <- top_station2 %>%
  group_by(start_station_name) %>%
  summarise(total_count = length(na.omit(start_station_name))) %>%
  arrange(desc(total_count)) %>%
  top_n(10)
```

```r
## Created viz for Top Ten stations for Casual Riders
ggplot(data = top_station_casual) +
  geom_col(mapping = aes(x = reorder(start_station_name, - total_count), y = total_count,
                         fill = start_station_name), position = "dodge") +
  labs(title = "Top Ten Start Stations for Casual", x = "start station name") +
  theme(axis.text.x = element_text(angle = 90)) +
  theme(legend.position = "none")
```
